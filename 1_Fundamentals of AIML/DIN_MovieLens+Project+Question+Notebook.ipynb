```python
#your code here
item_yr_genre = item.drop(['movie id','movie title','release date'], axis=1, inplace=False)
new_dataset = item_yr_genre.groupby("release year").sum()
new_dataset['total']= new_dataset.iloc[:, 0:].sum(axis=1)
new_dataset_per = new_dataset.div(new_dataset['total'], axis= 0) * 100

fig,ax = plt.subplots(figsize=(25, 25))
sns.set(font_scale=1.5)
sns.heatmap(new_dataset_per, annot=True, linewidths=0.05, fmt= '.2f',cmap="magma") # the color intensity is based on 
plt.show()
```

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#your code here

#Step1: Drop unnecessary columns, and to avoid impact on the original dataframe, create a new dataframe with desired data
movie_rating_df = Data.drop(['user id','timestamp'], axis=1, inplace=False)
#movie_rating_df.head()

#Step2: Count number of times a movie has been rated. Save grouped data into a dataframe
rating_count = movie_rating_df.groupby("movie id").count()

#Step3: Take the sum of all the ratings for each movie
rating_avg =  movie_rating_df.groupby("movie id").sum()

#Step4: Merge the dataframes obtained in step2 and step3
rating_count_avg = pd.merge(rating_count,rating_avg,how='inner',on='movie id')

#Step5: Calculate average. 
rating_count_avg["average rating"]=rating_count_avg["rating_y"] / rating_count_avg["rating_x"]

#Step6: Slice above dataframe to obtain movies that have been rated more than 100 times
rating_count_avg = rating_count_avg[rating_count_avg["rating_x"] > 100]

#Step7: Sort above dataframe according to "average rating" in descending order
rating_count_avg.sort_values(by="average rating",ascending=False,inplace=True)

#Step8: Get movie names
rating_count_avg = pd.merge(rating_count_avg,item,how='inner',on='movie id').round(2)

#Step9: Display top 25 movies
# Set following option to ensure that full movie title is diplayed. 
pd.set_option('max_colwidth', 400) 
rating_count_avg[["movie title", "average rating"]].iloc[:25,:]
```

### 10. Check for the validity of the below statements with respect to the data provided - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

**Please pay attention to what should be the denominator while calculating percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#your code here

#Step1: Merge Data with item to form a new dataframe Data_item. Use inner join to ensure maximum data accuracy.
Data_item = pd.merge(Data,item,how='inner',on='movie id')

#Step1: Merge Data_item with user to form a new dataframe Data_item_user. Use inner join to ensure maximum data accuracy.
Data_item_user = pd.merge(Data_item,user,how='inner',on='user id')

#Step3: Calculate total number of males and females in the dataset
cnt_g = (Data_item_user.groupby(["gender"])["gender"].count()).rename("Total")
#print(cnt_g)

#Step3: Calculate total number of males and females who rated movies in genre "Sci-Fi"
cnt_sci = (Data_item_user[Data_item_user["Sci-Fi"] == 1].groupby(["gender"])["gender"].count()).rename("SCI-FI")
#print(cnt_sci)

#Step4: Calculate total number of males and females who rated movies in genre "Romance"
cnt_rom = (Data_item_user[Data_item_user["Romance"] == 1].groupby(["gender"])["gender"].count()).rename("Romance")
#print(cnt_rom)

#Step5: Calculate total number of males and females who rated movies in genre "Drama"
cnt_dr = (Data_item_user[Data_item_user["Drama"] == 1].groupby(["gender"])["gender"].count()).rename("Drama")
#print(cnt_dr)

#Step6: Concatenate all the series obtained from steps 3-5 into a single dataframe
all_counts=pd.concat([cnt_g,cnt_sci,cnt_rom,cnt_dr],axis = 1, sort=True)

#Step7: Calculate %
all_counts['SCI-FI %']=all_counts["SCI-FI"]/all_counts["Total"]*100
all_counts['Romance %']=all_counts["Romance"]/all_counts["Total"]*100
all_counts['Drama %']=all_counts["Drama"]/all_counts["Total"]*100

#Step8: Display resultant dataframe and verify the statements
all_counts
```
